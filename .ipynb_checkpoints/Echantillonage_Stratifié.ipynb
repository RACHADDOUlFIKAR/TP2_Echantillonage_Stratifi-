# Avantages et Limites

## Exemple 1

```python
import numpy as np
from keras.datasets import mnist
```

```python
(x,y),(_,_)=mnist.load_data()
x = x.reshape(-1,28*28)/255.0
```

```python
y = (y==9)
```

# Exemple 2

```python
!pip install scikit-learn
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import StratifiedKFold
```

```python
skfolds = StratifiedKFold(n_splits=3)
splits = skfolds.split(x,y)
```

```python
for i,(train_index,test_index) in enumerate(splits):
    x_train = x[train_index]
    y_train = y[train_index]
    x_test = x[test_index]
    y_test = x[test_index]
    clf = SGDClassifier()
    clf.fit(x_train,y_train)
    y_pred = clf.predict(x_test)
    accuracy = np.mean(y_pred == y_test)
    print("[SPLIT%d]"%(i+1))
    print("Percentage of digit 9 in the original dataset:%.2f %%"%(np.mean(y==True)*100))
    print("Percentage of digit 9 in training set:%.2f %%"%(np.mean(y_train==True)*100))
    print("Percentage of digit 9 in test set:%.2f %%"%(np.mean(y_test==True)*100))
    print("Accuracy:%.4f"%accuracy)
```
